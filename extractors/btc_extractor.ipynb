```python
# Import librairies
import datetime
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Install yfinance and import the librairy
# https://www.geeksforgeeks.org/get-financial-data-from-yahoo-finance-with-python/
#pip install yfinance
import yfinance as yahooFinance
```

```python
# Here We are getting BTC financial information
# We need to pass BTC as argument for that
# https://finance.yahoo.com/quote/BTC-USD
GetBTCInformation = yahooFinance.Ticker("BTC-USD")
```

```python
# whole python dictionary is printed here
#print(GetBTCInformation.info)
```

```python
# Let us  get historical stock prices for Bitcoin 
# covering the past few years.
# max->maximum number of daily prices available 
# for Bitcoin.
# Valid options are 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y,
# 5y, 10y and ytd.

df = GetBTCInformation.history(period="max")

# startDate , as per our convenience we can modify
#startDate = datetime.datetime(2020, 1, 1)

# endDate , as per our convenience we can modify
#endDate = datetime.datetime(2021, 12, 31)

# pass the parameters as the taken dates for start and end
#df = GetBTCInformation.history(start=startDate, end=endDate)

df = df.drop(['Open', 'High', 'Low', 'Dividends', 'Stock Splits'], axis=1)

df.tail()
```

```python
# Compute volatility
TRADING_DAYS = 365

df['Log Returns'] = np.log(df['Close']/df['Close'].shift(1))
df['Log Returns'].fillna(0, inplace=True)

df['Annualized Volatility'] = df['Log Returns'].rolling(window=TRADING_DAYS).std() * np.sqrt(TRADING_DAYS)

df.tail()
```

```python
# Plot log returns
fig = px.histogram(df, x="Log Returns", title='BTC Log Returns')
fig.show()
```

```python
# Plot annualized volatility
fig = px.line(df, y="Annualized Volatility", title='BTC Annualized Volatility')
fig.show()
```
